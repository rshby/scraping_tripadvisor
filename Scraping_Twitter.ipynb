# Scraping Berdasarkan User Account
kita akan melakukan scraping untuk satu akun twitter, sebagai contoh saya gunakan akun twitter `@jokowi`.<br>
kolom yang akan discraping adalah sebagai berikut

```python
username = 'jokowi' # nama username akun twitter
max_tweets = 150  # jumlah data yang akan discraping
 
tweets = tweepy.Cursor(api.user_timeline, id=username, tweet_mode='extended').items(max_tweets)

# tuliskan parameter(kolom) apa saja yang akan discraping
tweets_list = [[tweet.full_text, tweet.created_at, tweet.id_str, tweet.user.screen_name, tweet.retweet_count, tweet.favorite_count, tweet.source] for tweet in tweets]
 
# buat DataFrame untuk menampung hasil
df_account = pd.DataFrame(tweets_list)
df_account.columns = ["text", "created_at", "id", "username", "retweet_count", "like_count", "source"]
```

# Scraping Berdasarkan keyword
kali ini kita akan melakukan scraping berdasarkan keyword tertentu. untuk contoh kali ini kita akan melakukan scraping data twitter berdasarkan keyword `Vaksin jogja`.<br>
daftar kolom yang ada adalah sebagai berikut

```python
text_query = 'Vaksin Jogja'  # keyword
max_tweets = 150  # jumlah data yang discraping
 
tweets = tweepy.Cursor(api.search,q=text_query).items(max_tweets)
 
# tuliskan parameter(kolom) apa saja yang akan discraping
tweets_list = [[tweet.text, tweet.created_at, tweet.id_str, tweet.user.name, tweet.user.screen_name, tweet.user.id_str, tweet.user.location, tweet.user.url, tweet.user.description, tweet.user.verified, tweet.user.followers_count, tweet.user.friends_count, tweet.user.favourites_count, tweet.user.statuses_count, tweet.user.created_at, tweet.user.profile_image_url_https] for tweet in tweets]

# simpan hasil scraping ke dalam DataFrame
df_keyword = pd.DataFrame(tweets_list)
df_keyword.columns = ["text", "created_at", "tweet_id", "user_profile_name", "username", "user_id", "user_location", "user_url", "user_desc", "user_verified", "user_followers_count", "user_friend_counts", "user_favourites_count", "user_statuses_count", "user_created_at", "user_profile_image_url"]
```
